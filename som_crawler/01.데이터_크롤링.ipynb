```python
import os
import re  ## 정규표현식 library import
import uuid
import datetime
import urllib.request
import json
import random

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
```

```python
## set_chrome_driver()
## => webdriver 준비 (기본 : 스텔스 모드, 백그라운드에서 크롤러 실행)

def set_chrome_driver():
    options = webdriver.ChromeOptions()
    ## 크롤러가 어떻게 작동하는지 직접 보고 싶다면, 아래 주석을 해제하고,  'options.add_argument('headless')' 줄을 주석처리
    ##options.add_argument('window-size=1920,1080')         
    options.add_argument('headless')
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    return driver
```

```python
def getAnimKind(driver):
    anim_code = {"개" : "417000", "고양이" : "422400"}
    
    kind_set = set()
    pk = 1
    
    for code in anim_code.items():        
        select_upkind =Select(driver.find_element("id","searchUpKindCd"))  ## 대분류에 대한 드롭다운 리스트(<select> 태그) 정보 가져옴 
        select_upkind.select_by_value(code[1]) ## 대분류에 대한 드롭다운 리스트에서, 개/고양이 선택
        
        select_kind=Select(driver.find_element("id","searchKindCd"))      ## 소분류에 대한 드롭다운 리스트 (<select> 태그) 정보 가져옴
        opt_list = select_kind.options        ## 소분류 드롭다운 리스트의 <option> 태그들을 List 로  가져옴      
        
        skip_l = ['전체','고양이']
        
        for opt in opt_list:
            if opt.text in skip_l:
                continue
            kind_set.add((code[0],opt.text))
            pk += 1                                   ## 각 <option> 태그들의 텍스트 콘텐츠(종 이름(리트리버, 비글...))들을 가져와 List 에 추가 
            
    return kind_set
```

```python
driver = set_chrome_driver()
driver.get(url='https://www.animal.go.kr/front/awtis/protection/protectionList.do?menuNo=1000000060')

kind_set = getAnimKind(driver)         # 개,고양이의 상세 종 이름들을 가져옴
```

```python
json_kind =[]

pk = 1

for k in kind_set:
    template ={
        'model': 'main.som_anim_category',
        'pk': pk,
        'fields': { 'ca_name1': '', 'ca_name2': ''}
    }
    template['fields']['ca_name1'] = k[0]
    template['fields']['ca_name2'] = k[1]
    json_kind.append(template)
    pk += 1
```

### 동물 정보들을 크롤링 해봅시다.

```python
URL = 'https://www.animal.go.kr/front/awtis/protection/protectionList.do?boardId=&page={}&pageSize=0&desertionNo=&menuNo=1000000060&searchSDate=2022-07-01&searchEDate=2022-07-31&searchUprCd=&searchOrgCd=&searchCareRegNo=&searchUpKindCd={}&searchKindCd=&searchSexCd='
## URL 템플릿 : Page (글 페이지 번호 : 1 ~ 4 까지만!)
##              searchUpKindCd (개, 고양이 코드 번호) 

IMG_PATH = '/Users/sinse/Computer_Science_conda/데이터_사이언스/anim_img/{}' # 이미지가 저장될 경로
```

```python
def img_download(path,img_url):    
    ext = img_url.split('.')[-1]
    filename = "%s.%s" % (uuid.uuid4(), ext)
    urllib.request.urlretrieve(img_url,path.format(filename))
    return filename
    
def getAnimInfo(driver,valid_hrs):
    anim_code = {'개' : '417000','고양이' : '422400'}
    
    info_pre_list = []
    
    for code in anim_code.items():
        for page in range(1,5):
            driver.get(url=URL.format(page,code[1]))    # 개 에 대한 4개 페이지, 고양이에 대한 4개 페이지 파싱 진행
            
            for i in range(10):                             # 각 페이지 내에 10 마리의 동물 정보 존재
                info_pre = {}
                info_link = driver.find_elements(By.CLASS_NAME, "moreBtn")[i]  ## i 번째 동물에 대한 링크 가져오기
                info_link.click()                                              ## 링크 클릭
                
                imgs = driver.find_elements(By.CSS_SELECTOR, ".publicPhotoArea>.publicPhoto>a>.photoArea")
                img_paths = [] 
            
                for i in range(len(imgs)):
                    img_url = imgs[i].get_attribute("src")
                    img_name = img_download(IMG_PATH ,img_url)
                    img_paths.append(img_name)
                    info_pre['이미지들'] = img_paths
            
                ths = driver.find_elements(By.CSS_SELECTOR, "table>tbody>tr>th")
                tds = driver.find_elements(By.CSS_SELECTOR, "table>tbody>tr>td")
            
                for i in range(len(ths)):
                    if ths[i].text in valid_hrs:
                        info_pre[ths[i].text] = tds[i].text
                
                info_pre_list.append(info_pre)
                driver.back()
            
    return info_pre_list
```

```python
anim_valid_hr = ['공고번호','품종', '성별', '나이/체중','발생장소', '접수일시', '중성화', '보호센터']

anim_pre = getAnimInfo(driver,anim_valid_hr)
```

```python
def parse_sex(string):
    codes ={'수컷' : 'M','암컷' : 'F' ,'미상' : 'M'}
    return codes[string]

def parse_neu(string):
    if string.strip() == '':
        string = '아니오'
    codes ={'예' : 'NE', '아니오' : 'IN'}
    return codes[string.strip()]

def parse_breed(string):
    patt = re.compile('\s*(\[개\]|\[고양이\])\s*([ㄱ-ㅎ|ㅏ-ㅣ|가-힣|\s]+)')
    m = patt.match(string)
    
    anim_kind1, anim_kind2 = '', ''
    
    if m.group():    ## 개 match
        anim_kind1 = m.group(1)[1:-1]
        anim_kind2 = m.group(2).strip()

    return anim_kind1, anim_kind2


def parse_age_kg(string):
    patt2 = re.compile('\s*([1-9][0-9]*|0)\s*\(년생\)\s*/\s*(([1-9][0-9]*|0)(\.[0-9]*)?)\s*\(Kg\)\s*')
    m2 =patt2.match(string)
    b_year = int(m2.group(1))
    weight = round(float(m2.group(2)),1)
    
    anim_age = datetime.datetime.now().year - b_year
    anim_weight = str(weight)
    

    return anim_age, anim_weight

def find_breed_no(kind):
    default = {'개' : '믹스견', '고양이':'믹스묘'}
        
    if kind[1] not in [k[1] for k in kind_set]:
        kind = (kind[0], default[kind[0]])
        
    kind = next(item for item in json_kind if item['model'] == 'main.som_anim_category' and item['fields'] == dict([('ca_name1',kind[0]),('ca_name2',kind[1])]))
    return kind['pk']
```

```python
## valid_hds = ['공고번호','품종', '성별', '나이/체중','발생장소', '접수일시', '중성화', '보호센터']
infos_pre_list_c = []

for a in anim_pre:
    infos_pre_list_c.append(a.copy())
```

```python
for info_c in infos_pre_list_c:
    info_c['품종'] = find_breed_no(parse_breed(info_c['품종']))
    info_c['성별'] = parse_sex(info_c['성별']) 
    info_c['나이/체중'] = parse_age_kg(info_c['나이/체중'])
    info_c['중성화'] = parse_neu(info_c['중성화'])
    
    info_c['발생장소'] = info_c['발생장소'].strip() 
    info_c['접수일시'] = info_c['접수일시'].strip()
    info_c['보호센터'] = info_c['보호센터'].strip()
    info_c['공고번호'] = info_c['공고번호'].strip()

print(infos_pre_list_c)    
```

### 보호소의 정보들을 크롤링해봅시다.

```python
URL = 'https://www.animal.go.kr/front/awtis/institution/institutionList.do?boardId=shop&page=1&pageSize=0&regNo=&menuNo=1000000059&searchUprCd=&searchOrgCd=&searchText={}'

def random_phone_num_generator():
    first = str(random.randint(100, 999))
    second = str(random.randint(1, 888)).zfill(3)
    last = (str(random.randint(1, 9998)).zfill(4))
    
    while last in ['1111', '2222', '3333', '4444', '5555', '6666', '7777', '8888']:
        last = (str(random.randint(1, 9998)).zfill(4))
    return '{}{}{}'.format(first, second, last)


def parsePH(ph):
    patt = re.compile('\s*([0-9]{2,})\-([0-9]{2,})-([0-9]{2,})\s*')
    m = patt.match(ph)
    
    if m :
        return m.group(1)+m.group(2)+m.group(3)
    else :
        return random_phone_num_generator()



def getSheltInfo(driver, shltr_pk, shltr_names):
    
    valid_hrs = {'보호센터명': 'shelter_name','보호센터주소' : 'shelter_addr','전화번호':'shelter_hp'}
    
    json_shltr_prof = []
    
    for shltr in shltr_names:
        idx = list(shltr_names).index(shltr)
        
        template = {
            "model": "login.som_shelter_profile", 
            "pk": 1, 
            "fields": {
                "user": 1, 
                "shelter_name": '', 
                "shelter_addr": '', 
                "shelter_hp": '', 
                "shelter_link": '', 
                "shelter_img": 'imgs/default_shelter.png', 
                "shelter_valid": 'A'
             }
        }
        
        driver.get(url=URL.format(shltr))
        
        try:
            shelter = driver.find_element(By.CSS_SELECTOR, "table>tbody>tr>td>a")
        except NoSuchElementException as n:
            '' 
        else:
            shelter.click()
            
            ths = driver.find_elements(By.CSS_SELECTOR, "table>tbody>tr>th")
            tds = driver.find_elements(By.CSS_SELECTOR, "table>tbody>tr>td")
            
            for i in range(len(ths)):
                if ths[i].text in valid_hrs:
                    shltr_feat = tds[i].text.strip()
                    if ths[i].text == '전화번호':
                        shltr_feat = parsePH(shltr_feat)
                    template['fields'][valid_hrs[ths[i].text]] = shltr_feat
                    
            driver.back()
            
        finally:
            template['pk'] = idx+1      
            template['fields']['user'] = shltr_pk[idx]         # pk 5 의 User 부터 보호소 계정과 매핑된다고 가정
            template['fields']['shelter_name'] = shltr
            json_shltr_prof.append(template)
            
            
    return json_shltr_prof
    
    
```

### 보호소 동물 정보와 보호소 정보를 연결해봅시다.

```python
def find_shelter(shltr_name):
    if shltr_name not in shltr_names:
        shltr_name = list(shltr_names)[0]
        
    shltr_prof = next(item for item in json_shltr_prof if item['fields']['shelter_name'] == shltr_name )
    return shltr_prof['pk']
 
for a_info in infos_pre_list_c:
    a_info['보호센터'] = find_shelter(a_info['보호센터'])
```

```python
def bind_name(a_breed):
    a_kind = next(item for item in json_kind if item['pk'] == a_breed)
    
    if a_kind['fields']['ca_name1']=='고양이':
        return '냥이'
    if a_kind['fields']['ca_name1']=='개':
        return '강아지'
```

```python
json_anim_infos = []
pk = 1

for info in infos_pre_list_c :
    template = { 
        "model": "main.som_animal", 
        "pk": 0, 
        "fields": {
            "anim_code": "", 
            "anim_name": "", 
            "anim_sex": "", 
            "anim_breed": 0, 
            "anim_regdate": "", 
            "anim_weight": "", 
            "anim_age": 2, 
            "anim_found": "", 
            "anim_post_id": 0, 
            "anim_neuter": "", 
            "shelter_id": 0, 
            "in_state_id": None, 
            "anim_img1": "", 
            "anim_img2": "", 
            "anim_img3": "", 
            "anim_img4": ""
        }
    }
    
    template['pk'] = pk
    
    template['fields']['anim_code'] = info['공고번호']
    template['fields']['anim_name'] = bind_name(info['품종'])
    template['fields']['anim_sex'] = info['성별']
    template['fields']['anim_breed'] = info['품종']
    template['fields']['anim_regdate'] = info['접수일시']
    template['fields']['anim_age'] = info['나이/체중'][0]
    template['fields']['anim_weight'] = info['나이/체중'][1]
    template['fields']['anim_found'] = info['발생장소']
    template['fields']['anim_post_id'] = pk
    template['fields']['anim_neuter'] = info['중성화']
    template['fields']['shelter_id'] =info['보호센터']
    
    for idx in range(0,len(info['이미지들'])):
        template['fields']["anim_img"+str(idx+1)] = 'imgs/animal/'+info['이미지들'][idx-1]

    json_anim_infos.append(template)
    pk += 1
    
```
